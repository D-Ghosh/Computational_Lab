```python
import numpy.random as r
from matplotlib import pyplot as plt
```

```python
n = 1000
s = [r.randint(0, 1) for i in range(n)]

head = 0
tail = 0
for i in range(n):
    if s[i] == 0:
        head += 1
    else:
        tail += 1

print('Probability of head', head/n)
print('Probability of tail', tail/n)

plt.hist(s, bins=2, edgecolor='black')
plt.show()
```

```python
n = 1000
D1 = [r.randint(1, 6) for i in range(n)]
D2 = [r.randint(1, 6) for i in range(n)]

S = [D1[i] + D2[i] for i in range(n)]

plt.hist(S, edgecolor='black')
plt.show()
```

```python
n = 10000
D1 = [r.randint(1, 6) for i in range(n)]
D2 = [r.randint(1, 6) for i in range(n)]
D3 = [r.randint(1, 6) for i in range(n)]
D4 = [r.randint(1, 6) for i in range(n)]


M = [D1[i] + D2[i] + D3[i] + D4[i] for i in range(n)]

plt.hist(M, edgecolor='black', log=False)
plt.show()
```
