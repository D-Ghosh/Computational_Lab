```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    a = -2.2067 * (10**(-12))
    b = 8.1 * (10**8)
    return a * (y**4 - b)

x, y0, n = [0, 500], 1200, 500
```

```python
def IVP_Solve(solver, x, y0, n):
    h = (x[1]-x[0]) / n
    X = np.linspace(x[0], x[1], n)
    Y = np.zeros(len(X))
    Y[0] = y0
    for n in range(len(X) - 1):
        Y[n + 1] = solver(X[n], Y[n], f, h)
    return X, Y

def euler(xn, yn, f, h):
    return yn + h * f(xn, yn)

def rk4(xn, yn, f, h):
    k1 = h * f(xn, yn)
    k2 = h * f(xn + 0.5 * h, yn + 0.5 * k1)
    k3 = h * f(xn + 0.5 * h, yn + 0.5 * k2)
    k4 = h * f(xn + h, yn + k3)
    return yn + (1/6) * (k1 + 2 * k2 + 2 * k3 + k4)
```

```python
X1, Y1 = IVP_Solve(euler, x, y0, n)
X2, Y2 = IVP_Solve(rk4, x, y0, n)
```

```python
plt.plot(X1, Y1, 'k-', label = 'Euler')
plt.plot(X2, Y2, '--', label = 'RK4')
plt.legend()
plt.show()
```

```python
print(f'Value of \\Theta(480s) is {Y1[180-1]:10.6f} K using Euler method')
print(f'Value of \\Theta(480s) is {Y2[480-1]:10.6f} K using RK4 method')
```
