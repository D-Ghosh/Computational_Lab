```python
from itertools import product

p = {}
for i in product(range(1,7), repeat=3):
    if sum(i) not in  p:
        p[sum(i)] = 0

    p[sum(i)] += 1

print(p)
```

```python
# Estimation of value of pi using Monte_Carlo_Simulation

n = 7  # power of tens
inside = sum(np.random.uniform(-1, 1, size=10**n)**2 + np.random.uniform(-1, 1, size=10**n)**2 <= 1)

pi = 4.0*(inside/10**n)
print('The value of pi is ', pi)
```

```python
# Integration using Monte_Carlo_Method

f = lambda x: np.sin(x)

def MCI(a, b, f):
    N = 1000000
    x = np.linspace(a, b, N)
    y_min = np.min(f(x))
    y_max = np.max(f(x)) + 1

    A = (b - a)*(y_max - y_min)           # Calculation of area
    x = np.random.uniform(a, b, N)
    y = np.random.uniform(y_min, y_max, N)
    n = np.sum([abs(y) < abs(f(x))])

    return A*(n/N)

MCI(0, np.pi, f)
```

```python
def MCI_mod(a, b, f):
    N = 1000000
    x = np.random.uniform(a, b, N)
    return (b - a)*np.mean(f(x))

MCI_mod(0, np.pi, f)
```
