```python
import numpy as np

x = []
steps = 0
for _ in range(int(10e4)):
    for i in range(100):
        n = np.random.uniform(0, 1)
        if n < 0.5:
            steps += -1
        else:
            steps += 1
    x.append(steps)
    steps = 0

x = np.asarray(x)
np.mean(x)


    
```
