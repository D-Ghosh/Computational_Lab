```python
import numpy as np

def f(a, xi, c, m):
    return (a*xi + c) % m

def GR(x0, f, n):
    a, c, m = 57, 1, 256
    x = []
    y = []
    x.append(x0)
    for i in range(n):
        y.append(f(a, x[i], c, m))
        x.append(y[i])

    return (x, y)
```

```python
n = 4000
x, y = GR(10, f, n)

from matplotlib import pyplot as plt
x = np.arange(n)
plt.plot(x, y, '.')
plt.show()
```

```python
y = np.random.randint(0, n, size=n)
plt.plot(x, y, '.')
plt.show()
```

```python
import numpy as np

N = 1000
c = np.random.uniform(-1, 1, size=N)
d = np.random.uniform(-1, 1, size=N)

inside = np.sum([c**2 + d**2 <= 1])

pi = 4.0*(inside/N)
print('The value of pi:', pi)
```

```python
f = lambda x: np.sin(x)

a, b, N = 0, np.pi, 100000
x = np.random.uniform(a, b, size=N)
I = (b - a)*np.mean(f(x))
print('The value of integration:', I)
```

```python
a, b, c, d, N = 0, 1, 0, 1, 100000

def f(x, y):
    return x + y

np.random.seed(25241263)
x = np.random.uniform(a, b, size=N)

np.random.seed(2352)
y = np.random.uniform(c, d, size=N)

(1 - 0)*np.mean(f(x,y))
```

```python
a, b, c, d, s, N, I = 0, 1, 0, 1, 0, 100000, []

for _ in range(10):
    np.random.seed(123528)
    x = np.random.uniform(a, b, size=N)

    np.random.seed(12547853)
    y = np.random.uniform(c, d, size=N)

    for i in range(len(x)):
        s = s + f(x[i], y[i])

    I.append((a - b)*(c - d)*(s/len(x)))
np.average(I)
```
